### Observations

* While females spend forty cents more than males, males account for more than 82% of total revenue.

* 35-39 yr-olds spend the most per person at $4.76 per person but the most revenue is generated by 20-24 yr-olds.

* The top five spenders made 5 or fewer purchases, totaling less than $20 per spender, indicating that the way to increase revenue is to increase number of players rather than increase purchases-per-player.

```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# File to Load (Remember to Change These)
file_to_load = "C:/Users/Nik/Documents/LaunchCoderGirlSTL/Data-Science-Assignments/course_material/week_04/homework/Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
#players_col = pd.DataFrame(purchase_data["SN"].unique())
#Total_Players = pd.DataFrame(players_col.count())
#Total_Players.columns = ['Total Players']
#Total_Players

players_col = pd.DataFrame([purchase_data['SN'].nunique()])
players_col.columns = ['Total Players']
players_col.index = ['']
players_col
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
purchase_data['Revenue'] = purchase_data['Price'] 
Summary = pd.DataFrame(purchase_data.agg({'Item ID':pd.Series.nunique, 'Price':np.mean, 'Purchase ID': pd.Series.count, 'Revenue':np.sum}))
Summary_Table = Summary.transpose()
Summary_Table.index = ['']
Summary_Table.columns = ['Number of Unique Items', 'Average Purchase Price', 'Total Number of Purchases', 'Total Revenue'] 
Summary_Table['Number of Unique Items'] = Summary_Table['Number of Unique Items'].map('{:,.0f}'.format)
Summary_Table['Average Purchase Price'] = Summary_Table['Average Purchase Price'].map('${:,.2f}'.format)
Summary_Table['Total Number of Purchases'] = Summary_Table['Total Number of Purchases'].map('{:,.0f}'.format)
Summary_Table['Total Revenue'] = Summary_Table['Total Revenue'].map('${:,.2f}'.format)
Summary_Table.T
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
Gender_Demographics = pd.DataFrame(purchase_data.groupby('Gender')['SN'].nunique())
Gender_Demographics['All Players'] = Gender_Demographics['SN'].sum()
Gender_Demographics['Percentage of Players'] = Gender_Demographics['SN'] / Gender_Demographics['All Players']* 100
Gender_Demographics.columns = ['Total Count','All Players','Percentage of Players']
Gender_Demographics['Percentage of Players'] = Gender_Demographics['Percentage of Players'].map('{:,.2f}%'.format)
Gender_Demographics[['Total Count','Percentage of Players']]

```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
Purchasing_by_Gender = purchase_data.groupby('Gender').agg({'SN':pd.Series.nunique,'Purchase ID':pd.Series.nunique, 'Price':np.mean, 'Revenue': np.sum,})
Purchasing_by_Gender['Avg Total Purchase per Person'] = Purchasing_by_Gender['Revenue']/Purchasing_by_Gender['SN']
Purchasing_by_Gender.columns = ['SN','Purchase Count','Average Purchase Price','Total Purchase Value','Avg Total Purchase per Person']
Purchasing_by_Gender['Avg Total Purchase per Person'] = Purchasing_by_Gender['Avg Total Purchase per Person'].map('${:,.2f}'.format)
Purchasing_by_Gender['Total Purchase Value'] = Purchasing_by_Gender['Total Purchase Value'].map('${:,.2f}'.format)
Purchasing_by_Gender['Average Purchase Price'] = Purchasing_by_Gender['Average Purchase Price'].map('${:,.2f}'.format)
Purchasing_by_Gender[['Purchase Count','Average Purchase Price','Total Purchase Value','Avg Total Purchase per Person']]
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
purchase_data['Age Group'] = pd.cut(x=purchase_data['Age'], bins=[1, 10, 15, 20, 25, 30, 35, 40,100], right=False, labels=['<10', '10-14', '15-19','20-24','25-29','30-34','35-39','40+'])
Age_Demographic_Table = pd.DataFrame(purchase_data.groupby('Age Group')['SN'].nunique())
Age_Demographic_Table['All Players'] = Age_Demographic_Table['SN'].sum()
Age_Demographic_Table['Percentage of Players'] = Age_Demographic_Table['SN'] / Age_Demographic_Table['All Players']* 100
Age_Demographic_Table['Percentage of Players'] = Age_Demographic_Table['Percentage of Players'].map('{:,.2f}%'.format)
Age_Demographic_Table.columns = ['Total Count','All Players','Percentage of Players']
Age_Demographic_Table[['Total Count','Percentage of Players']]
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_data['Age Group'] = pd.cut(x=purchase_data['Age'], bins=[1, 10, 15, 20, 25, 30, 35, 40,100], right=False, labels=['<10', '10-14', '15-19','20-24','25-29','30-34','35-39','40+'])
Age_Demographic_Table = purchase_data.groupby('Age Group').agg({'SN':pd.Series.nunique,'Purchase ID':pd.Series.nunique, 'Price':np.mean, 'Revenue': np.sum,})
Age_Demographic_Table['Avg Total Purchase per Person'] = Age_Demographic_Table['Revenue']/Age_Demographic_Table['SN']
Age_Demographic_Table.columns = ['SN','Purchase Count','Average Purchase Price','Total Purchase Value','Avg Total Purchase per Person']
Age_Demographic_Table['Avg Total Purchase per Person'] = Age_Demographic_Table['Avg Total Purchase per Person'].map('${:,.2f}'.format)
Age_Demographic_Table['Total Purchase Value'] = Age_Demographic_Table['Total Purchase Value'].map('${:,.2f}'.format)
Age_Demographic_Table['Average Purchase Price'] = Age_Demographic_Table['Average Purchase Price'].map('${:,.2f}'.format)
Age_Demographic_Table[['Purchase Count','Average Purchase Price','Total Purchase Value','Avg Total Purchase per Person']]
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
Top_Spenders_Table = purchase_data.groupby('SN').agg({'Purchase ID':pd.Series.nunique, 'Price':np.mean, 'Revenue': np.sum,})
Top_Spenders_Table.columns = ['Purchase Count', 'Average Purchase Price', 'Total Purchase Value']
Top_Spenders_Table['Total Purchase Value'] = Top_Spenders_Table['Total Purchase Value'].map('${:,.2f}'.format)
Top_Spenders_Table['Average Purchase Price'] = Top_Spenders_Table['Average Purchase Price'].map('${:,.2f}'.format)
Top_Spenders_Table.sort_values('Purchase Count', ascending=False).head()
```

# Matplotlib

```python
purchase_data[purchase_data['Gender']=='Female']['Revenue'].hist(bins=20,alpha=0.6)
purchase_data[purchase_data['Gender']=='Male']['Revenue'].hist(bins=20,alpha=0.3)
plt.legend(['Female','Male'])
plt.title('Revenue from Males vs. Females')
plt.xlabel('$ Spent')
plt.ylabel('Number of Individuals Making Purchases')
plt.show()
```
